# Brownian motion or Wiener process

```python
import math
import numpy as np
import numpy.random as npr
import scipy
import matplotlib.pyplot as plt
```

```python
def brownian_motion(dt=0.1, N=1000):
    
    #Initialize W(t) with zeros
    W = scipy.zeros(N+1)
    
    #We create N+1 timesteps: t = 0, 1, 2, ..., N
    t = scipy.linspace(0, N, N+1)
    
    #We have to use cumulative sum: on every step the additional value is drawn from a normal distribution
    #with mean 0 and variance dt*dt ... N(0, dt*dt)
    W[1:N+1] = scipy.cumsum(scipy.random.normal(0, dt, N)) #scipy.random.normal(0, dt, N) 
    #generates N points following normal distribution with mean = 0 and sd = dt
    #W est un array contenant N+1 nombres qui representent le prix du stock
    
    return t, W
```

```python
def plot_brownian_motion(t, W):
    plt.plot(t, W)
    plt.xlabel('Time (t)')
    plt.ylabel('Wiener process W(t)')
    plt.title('Wiener process')
    plt.show()
```

```python
t1, W1 = brownian_motion(0.0001, 1000) #low volatility
print(W1[0:10])
plot_brownian_motion(t1, W1)
```

```python
t2, W2 = brownian_motion(0, 1000) #volatility=0, so no change in value
print(W2[0:10])
plot_brownian_motion(t2, W2)
```

```python
t3, W3 = brownian_motion(100, 100) #high volatility, N=100 to see more clearly values on graph
print(W3[0:10])
plot_brownian_motion(t3, W3)
```
